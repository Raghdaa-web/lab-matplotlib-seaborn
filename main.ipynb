#### The data we will use in this challenge is:

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

#### Plot (x, y) and (x, z).
There are 2 ways of doing this. Do in both ways.

**Hint**: Check out the nrows, ncols and index arguments of subplots. 

Also, play around with the linewidth and style. Use the ones you're most happy with.

```python
# Method 1
# matplotlib hist
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='y = x * 2', color='blue')
plt.plot(x, z, label='z = x**2', color='red')

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Plot of y = x * 2 and z = x^2')
plt.legend()
plt.grid(True)
plt.show()
```

#### Use plt.subplots(nrows=1, ncols=2) to create the plot.

```python
# Method 2
fig, axs = plt.subplots(nrows=2, ncols=1, figsize=(8, 10))
axs[0].plot(x, y, label='y = x * 2', color='blue')
axs[0].set_title('Plot of y = x * 2')
axs[0].set_xlabel('x-axis')
axs[0].set_ylabel('y-axis')
axs[0].legend()
axs[1].plot(x, z, label='z = x^2', color='red')
axs[1].set_title('Plot of z = x^2')
axs[1].set_xlabel('x-axis')
axs[1].set_ylabel('y-axis')
axs[1].legend()

# Adjust layout to prevent overlap
plt.tight_layout()

# Display the subplots
plt.show()
```
